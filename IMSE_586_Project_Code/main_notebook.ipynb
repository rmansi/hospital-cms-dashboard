## 4. Cleaning & Feature Engineering

We:
- Remove exact duplicates.
- Select relevant variables for analysis.
- Convert numeric columns from strings (e.g., `"$12,345.67"` â `12345.67`).
- Create cost ratio features.


```python
# Remove duplicates just in case
merged_df = merged_df.drop_duplicates(subset=["Facility ID"]).reset_index(drop=True)

print("Merged dataset shape after dropping duplicates:", merged_df.shape)
```

```python
# Helper to clean dollar/percentage strings (if present)
def clean_money(s):
    if pd.isna(s):
        return np.nan
    return float(str(s).replace("$", "").replace(",", "").strip())

def clean_percent(s):
    if pd.isna(s):
        return np.nan
    return float(str(s).replace("%", "").strip())

# Example likely columns â adjust names if they are slightly different in your data
money_cols = [
    "Tot_Pymt_Amt",     # total payments
    "Tot_Mdcr_Pymt_Amt",    # medicare payments
    "Tot_Submtd_Cvrd_Chrg"  # covered charges
]

for col in money_cols:
    if col in merged_df.columns:
        merged_df[col] = merged_df[col].apply(clean_money)

# Some rating/percentage columns in General Info or HRRP
percent_cols = []
for col in ["Hospital overall rating", "Mortality national comparison",
            "Safety of care national comparison"]:
    if col in merged_df.columns and merged_df[col].dtype == "object":
        percent_cols.append(col)

for col in percent_cols:
    # Remove % and convert to numeric, setting errors to NaN
    merged_df[col] = pd.to_numeric(
        merged_df[col].astype(str).str.replace('%', '').str.strip(),
        errors='coerce'
    )

merged_df[money_cols].head()
```

```python
# Feature engineering: cost ratios, if cost columns exist
if set(money_cols).issubset(merged_df.columns):
    merged_df["Cost_Ratio_Total_vs_Medicare"] = (
        merged_df["Tot_Pymt_Amt"] / merged_df["Tot_Mdcr_Pymt_Amt"]
    )
    merged_df["Cost_Ratio_Charges_vs_Medicare"] = (
        merged_df["Tot_Submtd_Cvrd_Chrg"] / merged_df["Tot_Mdcr_Pymt_Amt"]
    )

merged_df.head()
```

```python
# Basic missingness summary for modeling-relevant columns
merged_df.isnull().mean().sort_values(ascending=False).head(20)
```

```python
# Simple handling: drop hospitals missing the target or key cost variables
target_col = "Excess Readmission Ratio"

model_df = merged_df.copy()
drop_cols = [target_col]

if set(money_cols).issubset(model_df.columns):
    drop_cols += money_cols

model_df = model_df.dropna(subset=drop_cols)
print("Shape after dropping NA rows for target + cost variables:", model_df.shape)
```

## 6. Modeling Setup

We build models to **predict the Excess Readmission Ratio** using hospital cost and quality metrics.

Models:
- Baseline **Multiple Linear Regression**
- **Random Forest Regressor**

Evaluation:
- Train/test split
- RMSE, MAE, RÂ²
- **Repeated cross-validation** (as suggested in instructor feedback)


```python
# 6.1 Select features

numeric_features = []
categorical_features = []

# Common numeric feature candidates (adjust to your actual columns)
for col in [
    "Avg_Tot_Pymt_Amt",
    "Avg_Mdcr_Pymt_Amt",
    "Avg_Submtd_Cvrd_Chrg_Amt",
    "Cost_Ratio_Total_vs_Medicare",
    "Cost_Ratio_Charges_vs_Medicare",
    "Predicted Readmission Rate",
    "Expected Readmission Rate"
]:
    if col in model_df.columns:
        numeric_features.append(col)

# Common categorical candidates
for col in [
    "Hospital Type",
    "Hospital Ownership",
    "State"
]:
    if col in model_df.columns:
        categorical_features.append(col)

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)
```

```python
# Define X, y
X = model_df[numeric_features + categorical_features]
y = model_df[target_col]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=586
)
X_train.shape, X_test.shape
```

```python
# 6.2 Preprocessing: scale numeric and one-hot encode categorical

numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

### 6.3 Baseline Model: Multiple Linear Regression with Cross-Validation


```python
linreg_model = Pipeline(steps=[
    ("preprocess", preprocess),
    ("regressor", LinearRegression())
])

# Repeated K-Fold cross-validation
cv = RepeatedKFold(n_splits=5, n_repeats=3, random_state=586)

cv_scores = cross_val_score(
    linreg_model, X_train, y_train,
    scoring="neg_root_mean_squared_error",
    cv=cv,
    n_jobs=-1
)

print("Linear Regression - CV RMSE (mean):", -cv_scores.mean())
print("Linear Regression - CV RMSE (std):", cv_scores.std())
```

```python
# Fit on full training data, evaluate on test set
linreg_model.fit(X_train, y_train)

y_pred_lr = linreg_model.predict(X_test)

# Calculate RMSE manually as the 'squared' argument is not supported in this scikit-learn version
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression Test RMSE:", rmse_lr)
print("Linear Regression Test MAE:", mae_lr)
print("Linear Regression Test R^2:", r2_lr)
```

### 6.4 Random Forest Regressor with Grid Search & Cross-Validation


```python
rf_model = Pipeline(steps=[
    ("preprocess", preprocess),
    ("regressor", RandomForestRegressor(random_state=586))
])

param_grid = {
    "regressor__n_estimators": [100, 200],
    "regressor__max_depth": [None, 10, 20],
    "regressor__min_samples_split": [2, 5]
}

grid_search = GridSearchCV(
    rf_model,
    param_grid,
    cv=5,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("Best RF params:", grid_search.best_params_)
print("Best CV RMSE:", -grid_search.best_score_)
```

```python
# Evaluate best RF on test set
best_rf = grid_search.best_estimator_
y_pred_rf = best_rf.predict(X_test)

# Calculate RMSE manually as the 'squared' argument is not supported in this scikit-learn version
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Test RMSE:", rmse_rf)
print("Random Forest Test MAE:", mae_rf)
print("Random Forest Test R^2:", r2_rf)
```

```python
# 1.4 Advanced missing data imputation
from sklearn.impute import KNNImputer, SimpleImputer

# Separate numeric and categorical for imputation
numeric_cols_impute = merged_df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols_impute = merged_df.select_dtypes(include=['object']).columns.tolist()

# For numeric: KNN imputation
numeric_imputer = KNNImputer(n_neighbors=5)
merged_df_numeric_imputed = pd.DataFrame(
    numeric_imputer.fit_transform(merged_df[numeric_cols_impute]),
    columns=numeric_cols_impute
)

# For categorical: mode imputation
categorical_imputer = SimpleImputer(strategy='most_frequent')
merged_df_categorical_imputed = pd.DataFrame(
    categorical_imputer.fit_transform(merged_df[categorical_cols_impute]),
    columns=categorical_cols_impute
)

# Combine back
merged_df_imputed = pd.concat(
    [merged_df_numeric_imputed, merged_df_categorical_imputed],
    axis=1
)

# Keep the original Facility ID
merged_df_imputed["Facility ID"] = merged_df["Facility ID"]

print("Missing values after imputation:")
print(merged_df_imputed.isnull().sum().head(10))
```

### Advanced Modeling with Gradient Boosting, Elastic Net, and Subgroup Analysis

```python
from sklearn.ensemble import GradientBoostingRegressor, StackingRegressor, VotingRegressor
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import cross_val_predict, KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 2.1 Update feature selection with new features
numeric_features = [
    "Tot_Pymt_Amt", "Tot_Mdcr_Pymt_Amt", "Tot_Submtd_Cvrd_Chrg",
    "Cost_Ratio_Total_vs_Medicare", "Cost_Ratio_Charges_vs_Medicare",
    "Predicted Readmission Rate", "Expected Readmission Rate",
    "Quality_Score"
]

# Add patient experience scores if available
for col in ["HCAHPS Answer Percent", "Patient Survey Star Rating",
            "Nurse Communication Linear Mean Score"]:
    if col in merged_df_imputed.columns:
        numeric_features.append(col)

categorical_features = [
    "Hospital Type", "Hospital Ownership", "State", "Region"
]

# Filter to columns that exist
numeric_features = [f for f in numeric_features if f in merged_df_imputed.columns]
categorical_features = [f for f in categorical_features if f in merged_df_imputed.columns]

print("Numeric features:", numeric_features)
print("Categorical features:", categorical_features)

# 2.2 Prepare data with imputed dataset
model_df = merged_df_imputed.dropna(subset=[target_col])
X = model_df[numeric_features + categorical_features]
y = model_df[target_col]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=586, stratify=model_df["Region"] if "Region" in model_df.columns else None
)

# 2.3 Enhanced preprocessing pipeline
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_features)
    ]
)

# 2.4 Define multiple advanced models
models = {
    "ElasticNet": Pipeline([
        ("preprocess", preprocess),
        ("regressor", ElasticNet(random_state=586, max_iter=10000))
    ]),
    "GradientBoosting": Pipeline([
        ("preprocess", preprocess),
        ("regressor", GradientBoostingRegressor(random_state=586, n_estimators=200))
    ]),
    "XGBoost": Pipeline([
        ("preprocess", preprocess),
        ("regressor", XGBRegressor(random_state=586, n_estimators=200, verbosity=0))
    ]),
    "LightGBM": Pipeline([
        ("preprocess", preprocess),
        ("regressor", LGBMRegressor(random_state=586, n_estimators=200, verbose=-1))
    ]),
    "SVR": Pipeline([
        ("preprocess", preprocess),
        ("regressor", SVR(kernel='rbf', C=1.0))
    ])
}

# 2.5 Hyperparameter tuning for each model
param_grids = {
    "ElasticNet": {
        "regressor__alpha": [0.01, 0.1, 1.0, 10],
        "regressor__l1_ratio": [0.1, 0.5, 0.9]
    },
    "GradientBoosting": {
        "regressor__n_estimators": [100, 200],
        "regressor__learning_rate": [0.01, 0.1, 0.2],
        "regressor__max_depth": [3, 5, 7]
    },
    "XGBoost": {
        "regressor__n_estimators": [100, 200],
        "regressor__max_depth": [3, 5, 7],
        "regressor__learning_rate": [0.01, 0.1]
    }
}

# 2.6 Train and evaluate all models
results = {}
best_models = {}

for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"Training {name}")
    print('='*50)

    if name in param_grids:
        # Perform grid search for models with defined param grids
        grid_search = GridSearchCV(
            model,
            param_grids[name],
            cv=5,
            scoring="neg_root_mean_squared_error",
            n_jobs=-1,
            verbose=0
        )
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
        best_models[name] = best_model
        print(f"Best params: {grid_search.best_params_}")
        print(f"Best CV RMSE: {-grid_search.best_score_:.4f}")
    else:
        # Use default parameters with cross-validation
        cv_scores = cross_val_score(
            model, X_train, y_train,
            scoring="neg_root_mean_squared_error",
            cv=5,
            n_jobs=-1
        )
        model.fit(X_train, y_train)
        best_models[name] = model
        print(f"CV RMSE: {-cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")

    # Evaluate on test set
    y_pred = best_models[name].predict(X_test)

    results[name] = {
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
        "MAE": mean_absolute_error(y_test, y_pred),
        "R2": r2_score(y_test, y_pred)
    }

    print(f"Test RMSE: {results[name]['RMSE']:.4f}")
    print(f"Test MAE: {results[name]['MAE']:.4f}")
    print(f"Test RÂ²: {results[name]['R2']:.4f}")

# 2.7 Create ensemble models
# Voting Regressor
voting_regressor = VotingRegressor([
    ("rf", best_models.get("RandomForest", rf_model)),
    ("gb", best_models.get("GradientBoosting")),
    ("xgb", best_models.get("XGBoost"))
])

voting_regressor.fit(X_train, y_train)
y_pred_voting = voting_regressor.predict(X_test)

results["VotingEnsemble"] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_voting)),
    "MAE": mean_absolute_error(y_test, y_pred_voting),
    "R2": r2_score(y_test, y_pred_voting)
}

# Stacking Regressor
stacking_regressor = StackingRegressor(
    estimators=[
        ("rf", best_models.get("RandomForest", rf_model)),
        ("gb", best_models.get("GradientBoosting")),
        ("xgb", best_models.get("XGBoost"))
    ],
    final_estimator=LinearRegression(),
    cv=5
)

stacking_regressor.fit(X_train, y_train)
y_pred_stacking = stacking_regressor.predict(X_test)

results["StackingEnsemble"] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred_stacking)),
    "MAE": mean_absolute_error(y_test, y_pred_stacking),
    "R2": r2_score(y_test, y_pred_stacking)
}

# 2.8 Compare all models
results_df = pd.DataFrame(results).T
print("\n" + "="*60)
print("MODEL COMPARISON")
print("="*60)
print(results_df.sort_values("RMSE"))

# 2.9 Subgroup Analysis
def analyze_subgroup(data, subgroup_col, model, X_cols):
    """Analyze model performance across subgroups"""
    subgroup_results = {}

    for subgroup in data[subgroup_col].unique():
        mask = data[subgroup_col] == subgroup
        if mask.sum() > 10:  # Only analyze subgroups with enough samples
            X_sub = X_cols[mask]
            y_sub = y[mask]

            if len(X_sub) > 0:
                y_pred_sub = model.predict(X_sub)
                subgroup_results[subgroup] = {
                    "n_samples": len(X_sub),
                    "RMSE": np.sqrt(mean_squared_error(y_sub, y_pred_sub)),
                    "MAE": mean_absolute_error(y_sub, y_pred_sub),
                    "R2": r2_score(y_sub, y_pred_sub),
                    "mean_actual": y_sub.mean(),
                    "mean_predicted": y_pred_sub.mean()
                }

    return pd.DataFrame(subgroup_results).T

# Analyze by region
print("\n" + "="*60)
print("SUBGROUP ANALYSIS BY REGION")
print("="*60)
if "Region" in X.columns:
    region_results = analyze_subgroup(
        pd.concat([X, y], axis=1).reset_index(drop=True),
        "Region",
        best_models["XGBoost"],  # Use best model
        X
    )
    print(region_results)

# Analyze by hospital type
print("\n" + "="*60)
print("SUBGROUP ANALYSIS BY HOSPITAL TYPE")
print("="*60)
if "Hospital Type" in X.columns:
    type_results = analyze_subgroup(
        pd.concat([X, y], axis=1).reset_index(drop=True),
        "Hospital Type",
        best_models["XGBoost"],
        X
    )
    print(type_results)
```
